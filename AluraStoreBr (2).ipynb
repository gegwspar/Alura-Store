### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
import matplotlib.pyplot as plt
faturamento = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f'O faturamento total da loja 1 foi de: {faturamento} ')
print(f'O faturamento total da loja 2 foi de: {faturamento_loja2} ')
print(f'O faturamento total da loja 3 foi de: {faturamento_loja3} ')
print(f'O faturamento total da loja 4 foi de: {faturamento_loja4} ')
```

# 2. Vendas por Categoria

```python
categoria_loja = loja.groupby('Categoria do Produto')['Produto'].count()
categoria_loja2 = loja2.groupby('Categoria do Produto')['Produto'].count()
categoria_loja3 = loja3.groupby('Categoria do Produto')['Produto'].count()
categoria_loja4 = loja4.groupby('Categoria do Produto')['Produto'].count()

print('Quantidade de produtos vendidos por categoria na Loja 1: ')
print(categoria_loja)
print('Quantidade de produtos vendidos por categoria na Loja 2:')
print(categoria_loja2)
print('Quantidade de produtos vendidos por categoria na Loja 3:')
print(categoria_loja3)
print('Quantidade de produtos vendidos por categoria na Loja 4:')
print(categoria_loja4)
```

# 3. Média de Avaliação das Lojas


```python
avaliacao_loja1 = loja['Avaliação da compra'].mean()
avaliacao_loja2 = loja2['Avaliação da compra'].mean()
avaliacao_loja3 = loja3['Avaliação da compra'].mean()
avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f'A média de avaliação da loja 1 é de: {avaliacao_loja1:.2f}')
print(f'A média de avaliação da loja 2 é de: {avaliacao_loja2:.2f}')
print(f'A média de avaliação da loja 3 é de: {avaliacao_loja3:.2f}')
print(f'A média de avaliação da loja 4 é de: {avaliacao_loja4:.2f}')
```

```python
avaliacoes = {
    'Loja 1': 3.98,
    'Loja 2': 4.04,
    'Loja 3': 4.05,
    'Loja 4': 4.00
}

plt.figure(figsize=(7,4))
plt.bar(avaliacoes.keys(), avaliacoes.values(), color='purple')
plt.title('Média de Avaliação das Lojas')
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação')
plt.ylim(3.8, 4.2)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()







```

# 4. Produtos Mais e Menos Vendidos

```python

loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

vendas = pd.concat([loja, loja2, loja3, loja4])

vendas_por_produto = vendas.groupby(['Loja', 'Produto']).size().reset_index(name='Quantidade')

mais_vendidos = []
menos_vendidos = []

for loja_nome in vendas_por_produto['Loja'].unique():
    dados_loja = vendas_por_produto[vendas_por_produto['Loja'] == loja_nome]
    mais = dados_loja.loc[dados_loja['Quantidade'].idxmax()]
    menos = dados_loja.loc[dados_loja['Quantidade'].idxmin()]
    mais_vendidos.append(mais)
    menos_vendidos.append(menos)

mais_vendidos = pd.DataFrame(mais_vendidos)
menos_vendidos = pd.DataFrame(menos_vendidos)

print(' PRODUTOS MAIS VENDIDOS POR LOJA:')
print(mais_vendidos[['Loja', 'Produto', 'Quantidade']])
print('\n PRODUTOS MENOS VENDIDOS POR LOJA:')
print(menos_vendidos[['Loja', 'Produto', 'Quantidade']])



```

```python
#  Produtos mais vendidos
plt.bar(mais_vendidos['Loja'], mais_vendidos['Quantidade'], color='lightgreen')
plt.title('Produto mais vendido por loja')
plt.xlabel('Loja')
plt.ylabel('Quantidade vendida')
plt.show()

#  Produtos menos vendidos
plt.bar(menos_vendidos['Loja'], menos_vendidos['Quantidade'], color='salmon')
plt.title('Produto menos vendido por loja')
plt.xlabel('Loja')
plt.ylabel('Quantidade vendida')
plt.show()
```

# 5. Frete Médio por Loja

```python
lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

for nome, df in lojas.items():
    media = df['Frete'].mean()
    print(f'A média de frete da {nome} é de: {media:.2f}')
```

```python
frete_medio = {
    'Loja 1': 34.69,
    'Loja 2': 33.62,
    'Loja 3': 33.67,
    'Loja 4': 31.28
}

plt.figure(figsize=(7,4))
plt.bar(frete_medio.keys(), frete_medio.values(), color='salmon')
plt.title('Frete Médio por Loja')
plt.xlabel('Lojas')
plt.ylabel('Valor Médio do Frete (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
